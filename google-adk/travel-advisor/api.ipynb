```python
import requests

BASEURL = "http://127.0.0.1:8000"
APPNAME = "travel_advisor"
USERID = "user123123"
SESSION_ID = (
    requests.post(f"{BASEURL}/apps/{APPNAME}/users/{USERID}/sessions")
    .json()
    .get("id", "")
)
```

```python
print(SESSION_ID)
```

```python
message = {
    "appName": APPNAME,
    "userId": USERID,
    "sessionId": SESSION_ID,
    "newMessage": {
        "parts": [
            {"text": "I am Korean"},
        ],
        "role": "user",
    },
}

response = requests.post(f"{BASEURL}/run", json=message)

data = response.json()

for event in data:
    content = event.get("content")
    parts = content.get("parts")
    for part in parts:
        function_call = part.get("functionCall")
        if function_call:
            print(function_call.get("name"))
        text = part.get("text")
        if text:
            print(text)
        print("===" * 60)
    print("*" * 30)
```

```python
import sseclient
import json

message = {
    "appName": APPNAME,
    "userId": USERID,
    "sessionId": SESSION_ID,
    "newMessage": {
        "parts": [
            {"text": "I want to go to Berlin, write as long as you can."},
        ],
        "role": "user",
    },
    "streaming": True,
}

response = requests.post(f"{BASEURL}/run_sse", json=message, stream=True)


def event_stream(response):
    for line in response.iter_lines():
        if line:
            # SSE data lines start with "data: "
            if line.startswith(b"data: "):
                yield line[len(b"data: ") :].decode("utf-8")


for event_data in event_stream(response):
    data = json.loads(event_data)

    content = data.get("content")
    parts = content.get("parts")
    for part in parts:
        function_call = part.get("functionCall")
        if function_call:
            print(function_call.get("name"))
        text = part.get("text")
        if text:
            print(text)
        print("===" * 60)
    print("*" * 30)
```
